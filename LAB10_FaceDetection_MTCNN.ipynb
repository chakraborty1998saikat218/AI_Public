```python
def bounding_box(image):
    faces=detector.detect_faces(image)
    box=faces[0]['box']
    keypoint=faces[0]['keypoints']
    cv2.rectangle(image,(box[0],box[1]),(box[0]+box[2],box[0]+box[3]),(0,255,0),2)

    cv2.circle(image,(keypoint['left_eye']),2,(0,155,255),2)
  
    cv2.circle(image,(keypoint['right_eye']),2,(0,155,255),2)
  
    cv2.circle(image,(keypoint['nose']),2,(0,155,255),2)
  
    cv2.circle(image,(keypoint['mouth_left']),2,(0,155,255),2)
  
    cv2.circle(image,(keypoint['mouth_right']),2,(0,155,255),2)

    return image
```

```python
marked_image=bounding_box(img)
cv2_imshow(marked_image)
```
